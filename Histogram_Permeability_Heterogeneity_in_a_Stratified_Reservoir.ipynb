```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
x=[10, 50, 150, 75, 100, 1, 32]
y=['Layer A', 'Layer B', 'Layer C', 'Layer D', 'Layer E', 'Layer F', 'Layer D']
```

```
plt.style.use('default')
plt.figure(figsize=(10,5))
```

```
plt.barh(y,x)
plt.xlabel('Permeability(mD)')
plt.ylabel('Layers')
plt.title('Permeability Heterogeneity in a Stratified Reservoir')
plt.grid()
```
